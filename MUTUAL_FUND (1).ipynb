# **letâ€™s have a look at the stock price trends of all the companies in the data:**

# **Letâ€™s look at the companies with the highest risks for investing:**

```python
all_companies=mutual_fund.columns[1:]
volatile_Companies=mutual_fund[all_companies].std().sort_values(ascending=False)
volatile_Companies.head(10)
```

# **letâ€™s look at the companies with the highest growth rate for investing:**

```python
growth_rate=mutual_fund[all_companies].pct_change()*100
average_growth_all_companies=growth_rate.mean()
average_growth_all_companies=average_growth_all_companies.sort_values(ascending=False)
average_growth_all_companies.head(10)
```

# **Letâ€™s have a look at the companies with the highest return on investments**

```python
intial_price=mutual_fund[all_companies].iloc[0]
final_price=mutual_fund[all_companies].iloc[-1]
roi_all_companies=round(((final_price-intial_price)/intial_price)*100,2)
roi_all_companies=roi_all_companies.sort_values(ascending=False)
roi_all_companies.head(10)
```

# **Creating a Mutual Fund Plan Based on High ROI and Low Risk**

---



# **we can use a combination of ROI and volatility (standard deviation) metrics.The goal is to find companies that offer a high return on investment (ROI) but with low volatility to minimize risk**

# **Letâ€™s start by defining thresholds and selecting companies that meet the criteria of high ROI and low volatility**

```python
roi_treshold=roi_all_companies.median()
volatility_threshold=volatile_Companies.median()
selecting_companies=round(roi_all_companies[(roi_all_companies>roi_treshold) & (volatile_Companies<volatility_threshold)],2)
selecting_companies.sort_values(ascending=False).head(10)
```

# **To balance the investment between these companies, we can use an inverse volatility ratio for allocation. Companies with lower volatility will get a higher weight. Letâ€™s calculate the weight for each company**

```python
selected_volatility=volatile_Companies[selecting_companies.index]
inverse_volatility=1/selected_volatility
investment_ratios=round((inverse_volatility/inverse_volatility.sum())*100,2)
investment_ratios.sort_values(ascending=False).head(10)
```

# **Analyzing Our Mutual Fund Plan**

# **Now, letâ€™s analyze and compare our mutual fund plan by comparing it with the high-performing companies in the stock market. Letâ€™s start by comparing the risks in our mutual fund with the risk in the high growth companies.**

```python
top_growth_companies=average_growth_all_companies.sort_values(ascending=False).head(10)
risk_growth_companies=volatile_Companies[top_growth_companies.index]
risk_mutualfund_companies=volatile_Companies[selecting_companies.index]
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

try:
    # Create the figure and axes
    fig, ax = plt.subplots(figsize=(10, 7)) # You can adjust the figure size as needed

    # Add bars for Mutual Fund Companies
    ax.barh(
        risk_mutualfund_companies.index,
        risk_mutualfund_companies.values,
        color='blue',
        label='Mutual Fund Companies'
    )

    # Add bars for Growth Rate Companies
    ax.barh(
        risk_growth_companies.index,
        risk_growth_companies.values,
        color='green',
        alpha=0.7,  # Opacity
        label='Growth Rate Companies'
    )

    # Set titles and labels
    ax.set_title('Risk Comparison: Mutual Fund vs Growth Rate Companies')
    ax.set_xlabel('Volatility (Standard Deviation)')
    ax.set_ylabel('Companies')

    # Set legend
    ax.legend(title='Company Type')

    # Use 'plotly_white' equivalent background (optional)
    fig.patch.set_facecolor('white')
    ax.set_facecolor('white')
    ax.grid(axis='x', linestyle='--', alpha=0.7) # Add grid for better readability, similar to plotly_white

    plt.tight_layout() # Adjust layout to prevent labels from overlapping
    plt.show()

except NameError as e:
    print(f"Error: {e}")
    print("Please make sure 'risk_mutual_fund_companies' and 'risk_growth_rate_companies' are defined before running this plotting code.")
    print("They should be pandas Series where the index represents company names and the values represent volatility.")
```

# **letâ€™s compare the ROI of both the groups as well**

```python
expected_roi_mutual_fund=roi_all_companies[selecting_companies.index]
expected_roi_growth_rate=roi_all_companies[top_growth_companies.index]
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


fig, ax = plt.subplots(figsize=(10, 7))

ax.barh(
    expected_roi_mutual_fund.index,
    expected_roi_mutual_fund.values,
    color='blue',
    label='Mutual Fund Companies'
)

ax.barh(
    expected_roi_growth_rate.index,
    expected_roi_growth_rate.values,
    color='green',
    alpha=0.7,  # Opacity
    label='Growth Rate Companies'
)

ax.set_title('Expected ROI Comparison: Mutual Fund vs Growth Rate Companies')
ax.set_xlabel('Expected ROI (%)')
ax.set_ylabel('Companies')

ax.legend(title='Company Type')

fig.patch.set_facecolor('white')
ax.set_facecolor('white')
ax.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

**The comparison between the risk (volatility) and expected ROI for mutual fund companies (in blue) and growth rate companies (in green) shows a clear trade-off. Mutual fund companies offer lower volatility, meaning they are less risky, but also provide lower expected returns. In contrast, growth rate companies demonstrate higher volatility, indicating more risk, but they offer much higher potential returns, especially companies like Bajaj Auto and Bajaj Finserv. This highlights a common investment dilemma: lower risk comes with a lower reward, while higher risk could yield higher returns**.

**For long-term investments, the goal is typically to find companies that offer a balance of stable returns and manageable risk. The companies in our mutual fund exhibit low volatility, meaning they are less risky, and their moderate returns make them solid choices for long-term, stable growth. They are well-suited for conservative investors who want steady returns without significant fluctuations in value**

# **Calculating Expected Returns**

**Now, letâ€™s calculate the expected returns a person will get from our mutual fund if he/she invests â‚¹5000 every month.**

**To calculate the expected value a person will accumulate over 1 year, 3 years, 5 years, and 10 years through the mutual fund plan, we can follow these steps:**

1.**Assume the person is investing 5000 rupees every month.**

2.**Use the expected ROI from the mutual fund companies to simulate the growth over time.**

4.**Compute the compounded value of the investments for each period (1y, 3y, 5y, and 10y).**

5.**Visualize the accumulated value over these periods.**

```python
monthly_investment=5000 # Monthly investment in INR
years=[1,3,5,10]   # Investment periods (in years)
n=12      # Number of times interest is compounded per year (monthly)

avg_roi=expected_roi_mutual_fund.mean()/100    # Convert to decimal
def future_value(p,r,n,t):
  return p*(((1+r/n)**(n*t)-1)/(r/n))*(1+r/n)
future_values=[future_value(monthly_investment,avg_roi,n,t) for t in years]

```

```python
plt.figure(figsize=(10, 6))
plt.plot(years, future_values, marker='o', color='blue', linestyle='-')

# Annotate each point with its value
for i, value in enumerate(future_values):
    plt.text(years[i], value, f"â‚¹{int(value):,}", ha='center', va='bottom', fontsize=9)

# Titles and labels
plt.title("Expected Value of Investments of â‚¹ 5000 Per Month (Mutual Funds)", fontsize=14)
plt.xlabel("Investment Period (Years)")
plt.ylabel("Future Value (INR)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(years)
plt.tight_layout()
plt.show()
```

# **After 1 year, the accumulated value is around â‚¹62,000, and by 5 years, it grows to over â‚¹300,000. The long-term benefit is evident, with the investment growing to nearly â‚¹860,000 over 10 years, which emphasises the value of consistent investing and compounding over time for long-term investors.**
